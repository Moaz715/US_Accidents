```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
final_df = pd.read_csv('mycsvfile.csv')

final_df.head()
```

```python
weather_cols = ['Temperature(F)', 'Humidity(%)', 'Pressure(in)', 
                'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)']

for col in weather_cols:
    plt.figure(figsize=(10,6))
    sns.barplot(data=final_df, x='Start_Month', y=col, estimator='mean', ci=None, palette='Blues')
    plt.title(f"Average {col} by Month in 2023")
    plt.xlabel("Month")
    plt.ylabel(col)
    plt.show()
    
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=final_df, x='Start_Month', hue='Severity', palette='Set2')
plt.title("Accident Counts per Month in 2023 by Severity")
plt.xlabel("Month")
plt.ylabel("Number of Accidents")
plt.legend(title="Severity")
plt.show()
```

```python


plt.figure(figsize=(14,6))
sns.countplot(data=final_df, x="Location", order=final_df['Location'].value_counts().index[:20], palette="viridis")
plt.xticks(rotation=90)
plt.title("Top 20 Locations by Accident Count")
plt.xlabel("Location")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
final_df['Start_Time'] = pd.to_datetime(final_df['Start_Time'])
final_df['End_Time'] = pd.to_datetime(final_df['End_Time'])
```

```python
final_df['Start_Hour'] = final_df['Start_Time'].dt.hour
final_df['Day']        = final_df['Start_Time'].dt.day_name()
final_df['End_Hour']   = final_df['End_Time'].dt.hour

```

```python
final_df.tail()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=final_df, x='Day', hue='Severity', palette='Set2')
plt.title("Accident Counts per Day in 2023 by Severity")
plt.xlabel("Day")
plt.ylabel("Number of Accidents")
plt.legend(title="Severity")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=final_df, x='Start_Hour', hue='Severity', palette='Set2')
plt.title("Accident Counts per Day in 2023 by Severity")
plt.xlabel("Hour")
plt.ylabel("Number of Accidents")
plt.legend(title="Severity")
plt.show()
```

```python

days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']


for day in days:
    day_data = final_df[final_df['Day'] == day]

    
    accidents_per_hour = day_data['Start_Hour'].value_counts().sort_index()

   
    plt.figure(figsize=(10,5))
    accidents_per_hour.plot(kind='bar', color='salmon', edgecolor='black')
    plt.title(f"Number of Accidents per Hour on {day}")
    plt.xlabel("Hour of Day")
    plt.ylabel("Number of Accidents")
    plt.xticks(rotation=0)
    plt.show()
```

```python
import dash
from dash import dcc, html
import plotly.express as px



app = dash.Dash(__name__)


fig_month = px.histogram(
    final_df, 
    x="Start_Month", 
    color="Severity", 
    barmode="group", 
    title="Accident Counts per Month in 2023 by Severity", 
    labels={"Start_Month":"Month", "count":"Number of Accidents"},
    color_discrete_sequence=px.colors.qualitative.Set2
)


top20_locations = final_df['Location'].value_counts().index[:20]
fig_locations = px.histogram(
    final_df[final_df['Location'].isin(top20_locations)],
    x="Location", 
    title="Top 20 Locations by Accident Count", 
    labels={"Location":"Location", "count":"Number of Accidents"},
    color_discrete_sequence=["#440154"]
)
#fig_locations.update_xaxes(categoryorder="total descending")


fig_day = px.histogram(
    final_df, 
    x="Day", 
    color="Severity", 
    barmode="group", 
    title="Accident Counts per Day in 2023 by Severity", 
    labels={"Day":"Day", "count":"Number of Accidents"},
    color_discrete_sequence=px.colors.qualitative.Set2
)


fig_hour = px.histogram(
    final_df, 
    x="Start_Hour", 
    color="Severity", 
    barmode="group", 
    title="Accident Counts per Hour in 2023 by Severity", 
    labels={"Start_Hour":"Hour", "count":"Number of Accidents"},
    color_discrete_sequence=px.colors.qualitative.Set2
)


days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
fig_day_hour = px.histogram(
    final_df, 
    x="Start_Hour", 
    facet_col="Day", 
    category_orders={"Day":days},
    title="Number of Accidents per Hour by Day of Week",
    labels={"Start_Hour":"Hour of Day", "count":"Number of Accidents"},
    color_discrete_sequence=["salmon"]
)
fig_day_hour.update_xaxes(matches=None, showticklabels=True)


app.layout = html.Div([
    html.H1("Accident Data Dashboard", style={"textAlign":"center"}),

    dcc.Tabs([
        dcc.Tab(label="Monthly Trends", children=[
            dcc.Graph(figure=fig_month)
        ]),
        dcc.Tab(label="Top 20 Locations", children=[
            dcc.Graph(figure=fig_locations)
        ]),
        dcc.Tab(label="Daily Trends", children=[
            dcc.Graph(figure=fig_day)
        ]),
        dcc.Tab(label="Hourly Trends", children=[
            dcc.Graph(figure=fig_hour)
        ]),
        dcc.Tab(label="Hourly Trends by Day", children=[
            dcc.Graph(figure=fig_day_hour)
        ])
    ])
])

if __name__ == "__main__":
    app.run(debug=True, port = 8050)
```
